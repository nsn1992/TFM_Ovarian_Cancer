```python
############# EXTRACT AND UNIFY INFORMATION OF RAW COUNTS OF TILS ###########################
# Import libraries to work
import pandas as pd
import numpy as np

############# LP samples ################
# Read the excel file
file1 = '/home/vant/TFM/Raw counts (no scores)/Results_ALEX_all.xlsx'

# Extract sheets with useful information
hoja2 = pd.read_excel(file1, sheet_name=0)  # First sheet (index starts at 0)
hoja4 = pd.read_excel(file1, sheet_name=1)  # Second sheet

# Verify column names to identify 'sample' and relevant count columns
print(hoja2.columns)
print(hoja4.columns)

# Select 'sample' and 'count_ep_caleiras' columns from sheet 2
df_hoja2 = hoja2[['sample', 'count_ep_caleiras']]

# Select 'sample' and 'count_tu_caleiras' columns from sheet 4
df_hoja4 = hoja4[['sample', 'count_tu_caleiras']]

# Merge the two dataframes using the 'sample' column as the key, keeping all data (outer join)
df_final = pd.merge(df_hoja2, df_hoja4, on='sample', how='outer')

# Sort the dataframe by the 'sample' column to align better the data
df_final_sorted = df_final.sort_values(by='sample')

# Drop any rows where 'sample' is completely NaN (empty rows at the bottom)
df_final_cleaned = df_final_sorted.dropna(subset=['sample'])
df_final_cleaned.columns=['Sample','Count_ep','Count_tu']

# Display the cleaned dataframe
print(df_final_cleaned)

# Save the cleaned dataframe to a new Excel file
#df_final_cleaned.to_excel("Resultado_alex_cleaned.xlsx", index=False)
```

```python
########## MDA samples #################

file2='/home/vant/TFM/Raw counts (no scores)/Results_MDA.xlsx'

# Read sheets 3 and 5
sheet3 = pd.read_excel(file2, sheet_name=2)  # Third sheet (index 2)
sheet5 = pd.read_excel(file2, sheet_name=4)  # Fifth sheet (index 4)

# Select relevant columns 'Sample' and 'Count_ep' from sheet 3
df_sheet3 = sheet3[['Sample', 'Count_ep']]

# Select relevant columns 'Sample' and 'Count_tu' from sheet 5
df_sheet5 = sheet5[['Sample', 'Count_tu']]

# Merge the dataframes based on 'Sample' using an outer join to include all rows
df_merged = pd.merge(df_sheet3, df_sheet5, on='Sample', how='outer')

# Strip leading/trailing spaces from the 'Sample' column
df_merged['Sample'] = df_merged['Sample'].str.strip()

# Convert 'Count_ep' and 'Count_tu' to numeric values, managing non-numeric data
df_merged['Count_ep'] = pd.to_numeric(df_merged['Count_ep'], errors='coerce')
df_merged['Count_tu'] = pd.to_numeric(df_merged['Count_tu'], errors='coerce')

# Group by 'Sample' and get the maximum values of 'Count_ep' and 'Count_tu' (including 9)
df_final = df_merged.groupby('Sample').agg({
    'Count_ep': 'max',  # Calculate the maximum, including the value 9
    'Count_tu': 'max'   # Same for 'Count_tu'
}).reset_index()

# Filter rows where the 'Sample' column starts with 'MDA'
df_final_filtered = df_final[df_final['Sample'].str.startswith('MDA')]

# Display the final cleaned dataframe
print(df_final_filtered)

# Save the final dataframe to a new Excel file
#df_final_filtered.to_excel('filtered_results_max_mda.xlsx', index=False)
```

```python
# Create a list with dataframes to later unify them
dataframes = [df_final_cleaned, df_final_filtered, df_final1_filtered, df_final2_filtered, df_final3, df_final4_filtered, df_final5]
# Create a list 
dfs = []

# Proccess every DataFrame
for df in dataframes:
    # Select columns of interest
    if 'Sample' in df.columns and 'Count_ep' in df.columns and 'Count_tu' in df.columns:
        # Seleccionar solo las columnas de interés
        dfs.append(df[['Sample', 'Count_ep', 'Count_tu']])

# Concatenate dataframes in only one
df_combined = pd.concat(dfs, ignore_index=True)
df_combined.shape
```

```python
# Eliminate rows in which both columns 'Count_ep' and 'Count_tu' are NaN
df_cleaned = df_combined.dropna(subset=['Count_ep', 'Count_tu'], how='all')

# Convert 'Count_ep' and 'Count_tu' columns to numeric using .loc and.tonumeric
df_cleaned.loc[:, 'Count_ep'] = pd.to_numeric(df_cleaned['Count_ep'], errors='coerce')
df_cleaned.loc[:, 'Count_tu'] = pd.to_numeric(df_cleaned['Count_tu'], errors='coerce')

# Show the cleaned dataframe
print(df_cleaned)

print("Longitud del dataframe final es:", len (df_cleaned))

# Save
df_cleaned.to_excel('/home/vant/TFM/Final_db/TILs_raw_counts_final.xlsx', index=False)
```

```python
# Read genomic + TILs scores database (Noelia Database complete.ods)
df_genomic = pd.read_excel('/home/vant/TFM/Noelia Database complete.ods', engine='odf')
df_genomic.columns = df_genomic.columns.str.strip()

# Read the Excel file (TILs_raw_counts_final.xlsx)
df_tils = pd.read_excel('/home/vant/TFM/Final_db/TILs_raw_counts_final.xlsx')

# Rename columns in df_tils to match df_genomic and for clarity
df_tils.rename(columns={'Sample': 'ID_CNIO', 'Count_ep': 'TILs_raw_ep', 'Count_tu': 'TILs_raw_tu'}, inplace=True)

# Merge the two DataFrames based on 'ID_CNIO' (outer join to keep all rows from df_genomic)
df_final = pd.merge(df_genomic, df_tils[['ID_CNIO', 'TILs_raw_ep', 'TILs_raw_tu']], on='ID_CNIO', how='left')

# Define the new column 'TILs_CNIO_info' based on 'TILs_raw_ep' and 'TILs_raw_tu'
df_final['TILs_CNIO_info'] = df_final.apply(
    lambda row: 'YES_complete' if pd.notna(row['TILs_raw_ep']) and pd.notna(row['TILs_raw_tu']) else
                'YES_only_ep' if pd.notna(row['TILs_raw_ep']) and pd.isna(row['TILs_raw_tu']) else
                'YES_only_tu' if pd.isna(row['TILs_raw_ep']) and pd.notna(row['TILs_raw_tu']) else
                'NO',
    axis=1
)

# Reorder columns to place 'TILs_CNIO_info' right after 'TILs_info'
cols = df_final.columns.tolist()
idx = cols.index('TILs counts') + 1  # Find the index of 'TILs_info'
# Insert 'TILs_CNIO_info' right after 'TILs_info'
cols.insert(idx, cols.pop(cols.index('TILs_CNIO_info')))
df_final = df_final[cols]

# Display the final DataFrame with the new column
print(df_final)

# Optionally, save the final DataFrame to an Excel file
df_final.to_excel('/home/vant/TFM/Final_db/Genomic_tils_final_database.xlsx', index=False)
```
