```python
## ### Preprocess of OVE records (Virgen del Rocio patients) ###############

# Import libraries and read clinical files
import pandas as pd
import numpy as np
# Read the first file with clinical data
ove_path='/home/vant/TFM/clinical data/Virgen del Rocio_OVE_series/TUMORES OVARIOS ESPORÁDICOS - ANTICUERPOS Y DATOS CLINICOS (TABLA BUENA 31-8-10)_datedeathlastvisitMaría.xlsx'
# To read the second sheet I must use the index 1 (0 is the first sheet and 1 the second one)
ove=pd.read_excel(ove_path, sheet_name=1,engine='openpyxl')
print(ove.head())
# Print information of the dataframe and length
print(ove.info())
print("The length/number of rows of the OVE dataframe is:", len(ove))
```

```python
# Read the second file with clinical data
ove_path2='/home/vant/TFM/clinical data/Virgen del Rocio_OVE_series/TUMORES OVARIOS ESPORÁDICOS - MORFOLOGÍA Y GRADACIÓN SILVERBERG 2000 Y MDACC 2004.xls'
ove2=pd.read_excel(ove_path2,engine='xlrd')
# Show first rows and print length of the datafram
print(ove2.head())
print("The length/number of rows of the OVE dataframe is:", len(ove2)) 
# Find the index of the first empty row
first_empty_row = ove2[ove2.isnull().all(axis=1)].index[0]
# Keep all rows until the first empty one
ove2 = ove2.iloc[:first_empty_row]
print("The length/number of rows of the OVE dataframe once eliminated the legends in the end is:", len(ove2))
```

```python
# Eliminate spaces in column names
ove.columns = ove.columns.str.strip()
ove2.columns = ove2.columns.str.strip()
print('Columns of OVE after eliminated spaces on the right and left:\n',ove.columns,'\n',ove2.columns)
# We see columns with the same names in the first file so we try to identify them
cols = ove.columns
duplicate_cols = cols[cols.duplicated()].tolist()
print("Columnas duplicadas:", duplicate_cols)
# As we can see, although there are no duplicate columns, we need to keep the .1 columns, which refer to the adjuvant treatment.
# The columns CISPLATINO.1 and TAXOL.1 contain only 0's, so they do not provide any useful information, and I will not select them.
```

```python
## Select columns of interest in every dataframe
columns_first=['CÓDIGO CNIO','FECHA NACIMIENTO','ESTADIO ( FIGO)','TRATAMIENTO NEOADYUVANTE', 'VOLUMEN TUMOR RESIDUAL (cm)', 'TRATAMIENTO ADYUVANTE',
               'CARBOPLATINO.1','CARBOPLATINO+TAXOL.1','CISPLATINO+TAXOL.1','OTROS.1','FECHA DE INTERVENCION',
               'DATE OF DEATH OR LAST VISIT','POSIBLE HEREDITARIO','OS','ESTADO ACTUAL','EXITUS']
columns_second = ['CÓDIGO CNIO', 'DIAGNÓSTICO', 'GRADO OMS', 'EDAD']
# Keep only the columns selected in every dataframe
ove=ove[columns_first]
ove2=ove2[columns_second]
# Check selected columns and length
print("Selected columns in the first file (ove):")
print(ove.columns)
print(len(ove))
print("\n Selected columns in the second file (ove2):")
print(ove2.columns)
print(len(ove2))
```

```python
# Columns to calculate overall survival ('OS_CNIO')
# Here we don´t have a diagnosis date, so we decide to use the surgery date ('FECHA DE INTERVENCION' column)
# Calculate OS. 
ove['FECHA DE INTERVENCION'] = pd.to_datetime(ove['FECHA DE INTERVENCION'],format='%d/%m/%Y')
ove['DATE OF DEATH OR LAST VISIT'] = pd.to_datetime(ove['DATE OF DEATH OR LAST VISIT'],format='%d/%m/%Y')
ove['OS_CNIO'] = (ove['DATE OF DEATH OR LAST VISIT'] - ove['FECHA DE INTERVENCION']).dt.days
# There is a OS in months provided in the file ('OS'column). We compare with our OS_CNIO.
ove['OS_CNIO_months'] = ove['OS_CNIO'] / 30.44  # Conversion to months

# Add a column indicating whether the date for calculating OS_CNIO is the date of surgery (except when we don´t know, NA)
# and compare OS_CNIO with the OS calculated by the hospital
print(ove[['CÓDIGO CNIO', 'DATE OF DEATH OR LAST VISIT', 'FECHA DE INTERVENCION', 'OS', 'OS_CNIO', 'OS_CNIO_months']])

# Create the new column 'DATE USED FOR OS' and add a legend in the row below
ove['DATE USED FOR OS'] = ove['FECHA DE INTERVENCION'].apply(lambda x: 'S' if pd.notnull(x) else 'NA')
# Check
print(ove[['CÓDIGO CNIO', 'DATE OF DEATH OR LAST VISIT', 'FECHA DE INTERVENCION','DATE USED FOR OS','OS', 'OS_CNIO', 'OS_CNIO_months']])
```

```python
# I check if the 'EDAD' column is the real age and not has been calculated from the surgery date ('FECHA DE INTERVENCION)
# Calculate 'AGE' using 'FECHA DE INTERVENCION' to compare with 'EDAD' column  

# Convert 'FECHA DE NACIMIENTO' to datetime, invalid entries will be converted to NaT (Not a Time)
ove['FECHA NACIMIENTO'] = pd.to_datetime(ove['FECHA NACIMIENTO'], format='%d/%m/%Y', errors='coerce')

# Convert 'FECHA DE INTERVENCION' to datetime, invalid entries will be converted to NaT (Not a Time)
ove['FECHA DE INTERVENCION'] = pd.to_datetime(ove['FECHA DE INTERVENCION'], format='%d/%m/%Y', errors='coerce')

# Calculate the age in years and round to the nearest whole number
ove['AGE'] = ((ove['FECHA DE INTERVENCION'] - ove['FECHA NACIMIENTO']).dt.days / 365).round()

# Optional: Check for rows where dates were invalid
#invalid_dates = ove[ove['FECHA NACIMIENTO'].isna() | ove['FECHA DE INTERVENCION'].isna()]
#print(invalid_dates)

# Check
print(ove[['CÓDIGO CNIO','FECHA DE INTERVENCION', 'FECHA NACIMIENTO', 'AGE']].head(40))
print(ove2[['CÓDIGO CNIO','EDAD']].head(40))
# After seeing the results I decide to keep the 'EDAD' column because ith has information of more samples. The differences
# between 'AGE'(our calculation) and 'EDAD' is that 'AGE' calculated for us is 1 plus year so it seemst to be estimated from the
# surgery date instead of the diagnosisa date. To indicate this I create an additional tag column 'DATE AGE'
ove['DATE_AGE'] = ove['FECHA DE INTERVENCION'].apply(lambda x: 'S' if pd.notnull(x) else 'NA')
```

```python
# Create FIGOL column with values 0 (localized- 1 to 8)) and 1 (advanced-9 to 13) as well as FIGOa column that returns:
# 1= IA, IB, IC, I(NOS); 2=IIA, IIB, II (NOS); 3=IIIA, IIIB, IIIC, III(NOS); 4= IV
# In the column 'ESTADIO ( FIGO)' there are values:  Ia,II,...
# We also create a FIGO STAGE column with the latin numbers from the original 'ESTADIO ( FIGO)'
import re
# I create a dictionary to later mapping
valid_roman_numerals = {
    'I': 0, 'II': 0, 'III': 1, 'IV': 1
}
# Function to extract the roman number and later obtain values for FIGOL column
def extract_roman_and_classify(stage):
    """Extrae el número romano completo de la cadena y clasifica como 0 (localizado) o 1 (avanzado) según el valor."""
    if isinstance(stage, str):
        # Keep only valid roman numbers (substitution of no valid values for empty chain '')
        normalized_stage = re.sub(r'[^IVXLCDM]', '', stage)

        # Regular expression to find the first valid roman number.If match is 'True' we map the values to obtain the values for FIGOL column
        match = re.search(r'(IV|I{1,3})', normalized_stage)
        if match:
            roman_numeral = match.group()
            if roman_numeral in valid_roman_numerals:
                return valid_roman_numerals[roman_numeral]
    return None

######## FIGOL column ##################
# Create FIGOL column from values in 'ESTADIO ( FIGO)'  
ove['FIGOL'] = ove['ESTADIO ( FIGO)'].apply(extract_roman_and_classify)

####### FIGO STAGE column ##########
# Define dictionary to map 'ESTADIO ( FIGO)' to 'FIGO STAGE' values
figo_stage_map = {
    'IA': 1, 'IB':2, 'IC':3, 'I':4,'IIA':5, 'IIB':6, 
    'II':8,'IIIA':9, 'IIIB':10, 'IIIC':11, 'III':12, 'IV':13
}
# To avoid errors I use upper() method before mapping (considering the data format in the original column)
ove['ESTADIO ( FIGO)']=ove['ESTADIO ( FIGO)'].str.upper()
ove['FIGO STAGE'] = ove['ESTADIO ( FIGO)'].map(figo_stage_map)

####### FIGOa column #############
# Create the 'FIGOa' column with the aggregated category 
figoa_map = {
    'IA': 1, 'IB':1, 'IC':1, 'I':1,'IIA':2, 'IIB':2, 
    'II':2,'IIIA':3, 'IIIB':3, 'IIIC':3, 'III':3, 'IV':4
}
ove['FIGOa']=ove['ESTADIO ( FIGO)'].map(figoa_map)
# Check
print(ove[['ESTADIO ( FIGO)','FIGO STAGE','FIGOa','FIGOL']]) # Las que me interesan son FIGO STAGE, FIGOa y FIGOL
```

```python
# The column that defines whether the patient is alive or deceased is called 'ESTADO ACTUAL'. 
# It contains a phrase indicating if the patient is alive, has been lost to follow-up, or if the status is unknown.
def determine_vital_status(status):
    """Determine the vital status depending on 'ESTADO ACTUAL' column."""
    if pd.isna(status) or status.strip() == '':
        return None  # NA for empty values
    if 'viva' in status.lower():
        return 0  # Alive
    elif 'perdido' in status.lower():
        return 0  # Also alive
    return None  # For other cases

# Apply the function to create the new column 'VITAL STATUS'
ove['VITAL STATUS'] = ove['ESTADO ACTUAL'].apply(determine_vital_status)
# I realise there is an additional column that allow us to fill some empty cells, the 'EXITUS' column. If the value in this column is 1 the
# patient is deceased,0 if it is alive and NA if we don't know the status

# Define a function to update NA/empty values in 'VITAL STATUS' according 'EXITUS' values
def update_vital_status(vital_status, exitus):
    if pd.isna(vital_status) or vital_status == '':
        if exitus == 1:
            return 1  # Assign 1 if'EXITUS' is 1
        else:
            return None  # Keep NA if 'EXITUS' is NA or other value
    return vital_status  # Keep the current value if not na or '': 0, 1 or 'LF'

# Apply the function to uptdate values in 'VITAL STATUS'
ove['VITAL STATUS'] = ove.apply(lambda row: update_vital_status(row['VITAL STATUS'], row['EXITUS']), axis=1)
# Check
print(ove[['ESTADO ACTUAL','EXITUS','VITAL STATUS']])
```

```python
# Create an additional column called HISTOLOGY based on the column DIAGNÓSTICO from the ove2 DataFrame
# Encode the values as follows: 0 = endometrioid, 1 = clear cell, 2 = mixed, 3 = others, NA = don't know.
# If the word "MIXTO" (mixed) is in the string values of DIAGNÓSTICO, it will be classified as mixed.
# First, remove leading and trailing spaces from the values in the DIAGNÓSTICO column.
ove2['DIAGNÓSTICO'] = ove2['DIAGNÓSTICO'].str.strip()

# Function to classify the histology based on the diagnosis string.
def classify_histology(diagnosis):
    if pd.isna(diagnosis):  # Check if the value is NaN (missing value)
        return np.nan
    elif diagnosis == 'CA ENDOMETRIOIDE':  # If the diagnosis is endometrioid cancer
        return 0
    elif diagnosis == 'CA CELS CLARAS':  # If the diagnosis is clear cell cancer
        return 1
    elif 'MIXTO' in diagnosis:  # If the diagnosis contains the word "MIXTO" (mixed)
        return 2
    else:  # If none of the above, classify as "others"
        return 3

# Create a new column HISTOLOGY by applying the classify_histology function to the DIAGNÓSTICO column
ove2['HISTOLOGY'] = ove2['DIAGNÓSTICO'].apply(classify_histology)

# Create another column HISTOLOGY_DETAIL. If the histology is classified as mixed (value 2 in the HISTOLOGY column),
# set the value of HISTOLOGY_DETAIL to the original DIAGNÓSTICO. Otherwise, set it to 'No mixed'.
ove2['HISTOLOGY_DETAIL'] = ove2.apply(
    lambda row: row['DIAGNÓSTICO'] if row['HISTOLOGY'] == 2 else 'No mixed', axis=1
)

# Select specific columns to display
columns_to_print = ['DIAGNÓSTICO', 'HISTOLOGY', 'HISTOLOGY_DETAIL']

# Print the first 77 rows of the selected columns
print(ove2[columns_to_print].head(77))
```

```python
# Add RESIDUALsD column
# Eliminar espacios al principio y al final de las cadenas
value_residuals= {
    '>0':1,'1':1,'2':1,'>2':1,'>2.5':1,'>4':1,'<1':1,'<2':1, '0':0
}
# Apply the function to create the RESIDUALsD column 
ove['VOLUMEN TUMOR RESIDUAL (cm)'] = ove['VOLUMEN TUMOR RESIDUAL (cm)'].str.strip()
ove['VOLUMEN TUMOR RESIDUAL (cm)'] = ove['VOLUMEN TUMOR RESIDUAL (cm)'].astype(str)
ove['RESIDUALsD'] = ove['VOLUMEN TUMOR RESIDUAL (cm)'].map(value_residuals)
print(ove[['VOLUMEN TUMOR RESIDUAL (cm)','RESIDUALsD']].head(45))
```

```python
# Combine dataframes using the common column 'CÓDIGO CNIO' 
df_combinado = pd.merge(ove, ove2, on='CÓDIGO CNIO', how='outer')
print(df_combinado)
print("La longitud del archivo combinado es:", len(df_combinado))
# Eliminate columns 'FECHA DE INTERVENCION','DATE OF DEATH OR LAST VISIT','ESTADO ACTUAL' y ESTADIO ( FIGO)
df_combinado = df_combinado.drop(columns=['FECHA DE INTERVENCION', 'DATE OF DEATH OR LAST VISIT','ESTADO ACTUAL','ESTADIO ( FIGO)'])

# Add legends below the header
leyendas = {
    'CÓDIGO CNIO': 'ID CNIO',
    'DIAGNÓSTICO': 'CA ENDOMETRIOIDE; CA CELS CLARAS; CA CELS CLARAS TARGETOIDE; CA MIXTO ENDOMETRIOIDE Y CELS CLARAS; CA MIXTO SEROSO Y CELS CLARAS; TUMOR CELS CLARAS BORDERLINE',
    'HISTOLOGY':'0=endometroid,1=clear cells,2=mixed,3=others,NA=no data',
    'HISTOLOGY_DETAIL':' Type of mixed cancer, No mixed, NA= unknown',
    'EDAD': 'years until the surgery',
    'DATE_AGE':'S=surgery date used for age, D=diagnosis date used for age, NA=unknown',
    'GRADO OMS':'1=well differentiated, 2=moderately differentiated, 3=poorly differentiated, NA=don´t know',
    'FIGO STAGE': '1=IA, 2=IB, 3=IC, 4=I(NOS), 5=IIA, 6=IIB, 7=IIC, 8=II(NOS), 9=IIIA, 10=IIIB, 11=IIIC, 12=III(NOS), 13=IV, NA=don´t know',
    'FIGOa': 'I,II,III,IV or NA=unknown',
    'FIGOL':'0=localized,1=advanced',
    'TRATAMIENTO NEOADYUVANTE': '0=NO;1=YES',
    'VOLUMEN TUMOR RESIDUAL (cm)': '0=no macroscopic disease;1=1cm;2=2cm;<1;<2=between 1 and 2 cm;NA=unknown',
    'RESIDUALsD':'0=No residual disease, 1=Yes residual disease',
    'TRATAMIENTO ADYUVANTE': '0=NO;1=YES',
    'CARBOPLATINO.1':'0=NO,1=YES',
    'CARBOPLATINO+TAXOL.1':'0=NO;1=YES',
    'CISPLATINO+TAXOL.1':'0=NO,1=YES',
    'OTROS.1':'Other adyuvant treatment',
    'OS_CNIO': 'Overall survival in days',
    'OS_CNIO_months': 'Overall survival in months',
    'OS': 'Overall survival in months from hospital',
    'DATE USED FOR OS':'D=diagnosis date used for OS; S=surgery date used for OS; NA=unknown',
    'POSIBLE HEREDITARIO':'0=NO;1=YES;NA=unknown',
    'VITAL STATUS':'0=alive,1=dead,LF:Lost follow-up'
}

# Create a legends dataframe
df_leyendas = pd.DataFrame([leyendas])
print(df_leyendas)

# Combine dataframes
df_final = pd.concat([df_leyendas, df_combinado], ignore_index=True)

# Save
df_final.to_excel('/home/vant/TFM/OVE_bbdd_filtered1.xlsx', index=False)

# See first rows of the dataframe
print(df_combinado[['CÓDIGO CNIO', 'OS', 'OS_CNIO', 'OS_CNIO_months']].head(10))
```
